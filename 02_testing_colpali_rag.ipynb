```python
import time
import logging

# Enable debug logging
logging.basicConfig(level=logging.INFO)

# Create an instance of the Companion
Companion = modal.Cls.lookup("colpali_rag", "Companion")
companion = Companion()

try:
    # First let's try with just the first page
    with open("LLM_Presentation (1).pdf", "rb") as f:
        pdf_bytes = f.read()

    print("Starting query process...")
    index_name = f"test_index_{int(time.time())}"
    
    results = companion.query.remote(
        doc=pdf_bytes,
        text="how to access open source llm?",
        index_name=index_name
    )
    
    if results:
        print(f"Got {len(results)} results!")
        for idx, image_bytes in enumerate(results):
            output_path = f"result_{idx}.png"
            with open(output_path, "wb") as f:
                f.write(image_bytes)
            print(f"Saved result to {output_path}")
    else:
        print("No results returned")

except Exception as e:
    print(f"Error occurred: {str(e)}")
    # Print the full error traceback for debugging
    import traceback
    print(traceback.format_exc())
```

```python
# Create an instance of the Companion
companion = Companion()

# First, index your document
try:
    with open("docs\Different Agent Frameworks.pdf", "rb") as f:
        pdf_bytes = f.read()
    
    doc_id = companion.index_document.remote(
        doc=pdf_bytes,
        doc_id="my_custom_id21"  # optional
    )
    print(f"Document indexed with ID: {doc_id}")

    # Now you can search multiple times without reloading the model
    results = companion.search.remote(
        query="what is autogen?",
        k=3
    )
    
    if results:
        print(f"Got {len(results)} results!")
        for idx, image_bytes in enumerate(results):
            output_path = f"result_{idx}.png"
            with open(output_path, "wb") as f:
                f.write(image_bytes)
            print(f"Saved result to {output_path}")
    else:
        print("No results found")

except Exception as e:
    print(f"Error: {str(e)}")
```
